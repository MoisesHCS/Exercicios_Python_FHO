```python
numero = int(input("Digite um número:"))
fatorial = 1
expressao = f"{numero}! = "
for i in range(numero, 1, -1):
  fatorial *= i
  expressao += f"{i} * "

expressao += "1"
print(expressao)
print(f"O fatorial de {numero} é: {fatorial}")
```

13. Escreva um programa que pergunte ao usuário quantos alunos têm na sala dele. Em
seguida, peça ao usuário para que entre com as notas de todos os alunos da sala, e, para
cada aluno, o programa imprime se ele está aprovado ou reprovado.

```python
alunos = int(input("Digite o número de alunos da sua sala:"))
for i in range(alunos):
  n1 = float(input("Digite a nota n1:"))
  n2 = float(input("Digite a nota n2:"))
  media = (n1 + 2*n2)/3
  print(f"média = {media}")
  if (media >= 5):
    print("APROVADO")
  elif (media >= 3):
    print("RE")
  else:
    print("REPROVADO")
```

14. Faça um programa que gere os n primeiros termos de uma PA. Dados como a

15. Faça um programa que seja capaz receber N valores fornecidos pelo usuário e exiba
na tela o valor da multiplicação dos valores pares.

```python
n = int(input("Digite o número de valores a serem fornecidos:"))
mult = 1
for i in range(n):
  valor = float(input("Digite o valor:"))
  if (valor % 2 == 0):
    mult = mult * valor
print(f"O valor da multiplicação dos valores pares é: {mult}")
```

17. Faça um programa que pergunte ao usuário em que ano ele nasceu e em que ano ele
está, e mostre de forma progressiva todos os anos que ele já viveu. Obs.: Valores
negativos não são aceitos.

```python
ano_nascimento = int(input("Digite o ano em que você nasceu:"))
ano_atual = int(input("Digite o ano em que você está:"))
if ano_nascimento < 0 or ano_atual < 0:
  print("Valores negativos não são aceitos")
else:
  for i in range(ano_nascimento, ano_atual + 1):
    print(i)
```

18. Faça um programa que receba o valor de início e de fim de um intervalo. Imprima
os valores do intervalo na tela, sendo que, será impresso “Fizz” para os múltiplos de 3,
“Ban” para os múltiplos de 5 e “FizzBan” para os múltiplos de 3 e 5 ao mesmo tempo.

```python
inicio = int(input("Digite o início do intervalo:"))
fim = int(input("Digite o fim do intervalo:"))
for i in range(inicio, fim + 1):
  if(i%3 == 0 and i%5 == 0):
    print("FizzBan")
  elif(i%3 == 0):
    print("Ban")
  elif(i%5 == 0):
    print("Ban")
  else:
    print(i)
```
