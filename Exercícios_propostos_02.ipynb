5)Um polígono convexo de n lados apresenta um número de diagonais (nd) diferente,
que pode ser calculado pela seguinte expressão. Faça um programa que leia o número de lados do polígono (n) e calcule o número de
diagonais diferentes do polígono. Realize o cálculo somente se o valor lido em n for maior que 2


```python
n = int(input("Digite o número de lados do polígono:"))
if (n > 2):
  print(f"O número de diagonais do polígono é {n*(n-3)/2}")
```

6)O índice de massa corpórea (IMC) é. No qual a altura é dada em metros e o peso é dado em quilogramas.
Considerando isso, faça um programa que calcule o IMC da pessoa, cujo valor ideal
está entre 18,5 e 24,9. O programa deve alertar o usuário.

```python
peso = float(input("Informe seu peso:"))
altura = float(input("Informe sua altura:"))
imc = float(peso/altura**2)
print(f"Seu imc é {imc}")
if (imc < 18.5):
  print("IMC ABAIXO DO IDEAL")
if (imc > 24.9):
  print("IMC ACIMA DO IDEAL")
```
