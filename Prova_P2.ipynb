1) Escreva um programa em Python que imprima os números de 1 a 50. No entanto, para os múltiplos
de 5, imprima "FHO" em vez do número.
Exemplo: 1 2 3 4 FHO 6 7 8 9 FHO ...

```python
for i in range(1, 51):
  if(i%5 == 0):
    print("FHO")
  else:
    print(i)
```

2) Escreva um programa que solicite ao usuário inserir o lado de 5 cubos. Utilizando uma estrutura de
repetição, calcule e exiba o volume de cada cubo. A fórmula do volume do cubo é dada por V = lado3.

```python
for i in range(5):
  l = float(input("Digite o valor do lado do cubo:"))
  V = l**3
  print(f"Volume = {V}")
```

3) Escreva um programa em Python que peça ao usuário informar o peso de 8 pessoas. Utilizando
uma estrutura de repetição, classifique e imprima na tela se cada pessoa é considerada Peso Leve (até
60 kg), Peso Médio (60 kg a 80 kg) ou Peso Pesado (acima de 80 kg).

```python
for i in range(8):
  P = float(input("Digite o peso da pessoa em kg:"))
  if(P < 60):
    print("Peso Leve")
  elif(P > 80):
    print("Peso Pesado")
  else:
    print("Peso Médio")
```

4) Faça um programa que imprima na tela os valores do intervalo de 10 a 1, retroativamente. Quando
o valor for zero, o programa deve imprimir na tela a palavra “FIM”.

```python
for i in range(10, 0, -1):
  print(i)
print("FIM")
```
